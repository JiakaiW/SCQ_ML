```python
import pandas as pd
import matplotlib.pyplot as plt
import argparse

def plot_loss_history(log_file, output_file):
    """
    Reads a training log CSV file and plots the training and test loss.

    Args:
        log_file (str): Path to the CSV log file.
        output_file (str): Path to save the output plot image.
    """
    try:
        # Read the data from the CSV file
        df = pd.read_csv(log_file)

        # Create the plot
        plt.figure(figsize=(10, 6))
        plt.plot(df['epoch'], df['train_loss'], label='Training Loss')
        plt.plot(df['epoch'], df['test_loss'], label='Test Loss')
        plt.axhline(y=min(df['test_loss']), color='r', linestyle='--', label='best test loss')
        # Set plot labels and title
        plt.xlabel('Epoch')
        plt.ylabel('Loss (MSE)')
        plt.title('Training and Test Loss Over Epochs')
        plt.legend()
        plt.grid(True, which="both", ls="--")
        plt.xlim(0,200)
        plt.ylim(1e-5,1)
        # Use a logarithmic scale for the y-axis
        plt.yscale('log')

        # Save the plot to a file
        plt.savefig(output_file, dpi=300)
        print(f"Plot saved to {output_file}")

        # Show the plot
        plt.show()

    except FileNotFoundError:
        print(f"Error: The file '{log_file}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")


plot_loss_history("../logs/training_log_graphformer.csv", "../loss_history.png") 
```
